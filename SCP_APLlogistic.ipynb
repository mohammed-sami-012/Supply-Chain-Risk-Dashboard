```python
# Shipping pressure index (scheduled days vs quantity)
SupplyChainAPL['Shipping Pressure Index'] = (
    SupplyChainAPL['Order Item Quantity'] / (SupplyChainAPL['Days for shipment (scheduled)'] + 1)
)
```

```python
# Order Complexity Score
qty_norm = (SupplyChainAPL['Order Item Quantity'] - SupplyChainAPL['Order Item Quantity'].min()) /\
           (SupplyChainAPL['Order Item Quantity'].max() - SupplyChainAPL['Order Item Quantity'].min())

price_norm = (SupplyChainAPL['Product Price'] - SupplyChainAPL['Product Price'].min()) /\
             (SupplyChainAPL['Product Price'].max() - SupplyChainAPL['Product Price'].min())

SupplyChainAPL['Order Complexity Score'] = (qty_norm + price_norm) / 2
```

```python
#Drop 'Delivery Status' (leaks into Late_delivery_risk target)
SupplyChainAPL = SupplyChainAPL.drop(columns=['Delivery Status'])
X = SupplyChainAPL.drop(columns=['Late_delivery_risk', 'Profitability Flag',
                                   'Days for shipping (real)', 'Days for shipment (scheduled)'])
y = SupplyChainAPL['Late_delivery_risk']
```

```python
#building risk flag
express_modes = ['Same Day', 'First Class']

SupplyChainAPL['Mode Risk Flag'] = SupplyChainAPL['Shipping Mode'].apply(
    lambda x: 1 if x in express_modes else 0
)
```

```python
SupplyChainAPL.groupby('Mode Risk Flag')['Late_delivery_risk'].mean()
```

```python
#One-hot encode low-cardinality columns
low_card_cols = ['Type', 'Customer Segment', 'Shipping Mode', 'Customer Country', 'Order Status']
SupplyChainAPL = pd.get_dummies(
    SupplyChainAPL,
    columns=low_card_cols,
    drop_first=True
)
```

```python
#Train/test split
from sklearn.model_selection import train_test_split

x = SupplyChainAPL.drop(columns=['Late_delivery_risk', 'Days for shipping (real)', 'Days for shipment (scheduled)'])
y = SupplyChainAPL['Late_delivery_risk']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# Regional Congestion Indicator
region_counts = x_train['Order Region'].value_counts()

x_train['Regional Congestion Indicator'] = x_train['Order Region'].map(region_counts)
x_test['Regional Congestion Indicator'] = x_test['Order Region'].map(region_counts)

x_test['Regional Congestion Indicator'] = x_test['Regional Congestion Indicator'].fillna(0)
```

```python
#Target-encode high-cardinality columns (fit on train, apply to test)
high_card_cols = ['Category Name', 'Department Name', 'Order Region', 'Product Name']

for col in high_card_cols:
    target_means = y_train.groupby(x_train[col]).mean()

    x_train[col + '_encoded'] = x_train[col].map(target_means)
    x_test[col + '_encoded'] = x_test[col].map(target_means)

    overall_means = y_train.mean()
    x_test[col + '_encoded'] = x_test[col + '_encoded'].fillna(overall_means)

    x_train.drop(columns=[col], inplace=True)
    x_test.drop(columns=[col], inplace=True)
```

```python
# fix the leftover Profitability Flag column
x_train = x_train.drop(columns=['Profitability Flag'])
x_test = x_test.drop(columns=['Profitability Flag'])
```

```python
# scale numerical features
from sklearn.preprocessing import StandardScaler

num_cols = ['Sales per customer',
    'Order Item Profit Ratio', 'Order Item Quantity', 'Sales', 'Order Item Total',
    'Order Profit Per Order', 'Product Price', 'Shipping Pressure Index',
    'Order Complexity Score', 'Regional Congestion Indicator',
    'Category Name_encoded', 'Department Name_encoded',
    'Order Region_encoded', 'Product Name_encoded'
]

scaler = StandardScaler()

x_train[num_cols] = scaler.fit_transform(x_train[num_cols])
x_test[num_cols] = scaler.transform(x_test[num_cols])
```

```python
#Address class imbalance (class weights)
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(class_weight='balanced', max_iter=1000, random_state=42)
model.fit(x_train, y_train)
```

```python
#classification report
from sklearn.metrics import classification_report

y_pred = model.predict(x_test)
print(classification_report(y_test, y_pred))
```

```python
# Coefficient interpretation
coefficients = pd.DataFrame({
    'Feature': x_train.columns,
    'Coefficient': model.coef_[0]
})

coefficients['Odds Ratio'] = np.exp(coefficients['Coefficient'])
coefficients = coefficients.sort_values('Coefficient', key=abs, ascending=False)

print(coefficients.head(15))
```

```python
#baseline's ROC-AUC
from sklearn.metrics import roc_auc_score

y_proba = model.predict_proba(x_test)[:, 1]
print('ROC-AUC:', roc_auc_score(y_test, y_proba))
```

```python
#random forest model
from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier(
    n_estimators=200,
    class_weight='balanced',
    random_state=42
)
rf_model.fit(x_train, y_train)
```

```python
#random forest model classification report
from sklearn.metrics import classification_report, roc_auc_score

y_pred_rf = rf_model.predict(x_test)
y_proba_rf = rf_model.predict_proba(x_test)[:, 1]

print(classification_report(y_test, y_pred_rf))
print('ROC-AUC:', roc_auc_score(y_test, y_proba_rf))
```

```python
#Feature importance
rf_importance = pd.DataFrame({
    'Feature': x_train.columns,
    'Importance': rf_model.feature_importances_
}).sort_values('Importance', ascending=False)

print(rf_importance.head(15))
```

```python
#baseline's ROC-AUC
from sklearn.metrics import roc_auc_score

y_proba_log = model.predict_proba(x_test)[:, 1]
print('Logistic Regression ROC-AUC:', roc_auc_score(y_test, y_proba_log))
```

```python
#XGBoost model
from xgboost import XGBClassifier

scale_pos_weight = (y_train == 0).sum() / (y_train == 1).sum()

xgb_model = XGBClassifier(
    n_estimators=200,
    scale_pos_weight=scale_pos_weight,
    random_state=42,
    eval_metric='logloss'
)
xgb_model.fit(x_train, y_train)
```

```python
#XGBoost model evaluate
y_pred_xgb = xgb_model.predict(x_test)
y_proba_xgb = xgb_model.predict_proba(x_test)[:, 1]

print(classification_report(y_test, y_pred_xgb))
print('ROC-AUC:', roc_auc_score(y_test, y_proba_xgb))
```

```python
#Feature importance
xgb_importance = pd.DataFrame({
    'Feature': x_train.columns,
    'Importance': xgb_model.feature_importances_
}).sort_values('Importance', ascending=False)

print(xgb_importance.head(15))
```

```python
#baseline's ROC-AUC
from sklearn.metrics import roc_auc_score

y_proba_log = model.predict_proba(x_test)[:, 1]
print('Logistic Regression ROC-AUC:', roc_auc_score(y_test, y_proba_log))
```

```python
#choseing best model ROC AUC score
best_proba = y_proba_xgb
```

```python
#Building risk probability table
risk_results = pd.DataFrame({
    'Late Delivery Probability': best_proba
}, index=x_test.index)

risk_results.head()
```

```python
#Classification of risk probability
def classify_risk(prob):
    if prob < 0.33:
        return 'Low Risk'
    elif prob < 0.66:
        return 'Medium Risk'
    else:
        return 'High Risk'

risk_results['Risk Category'] = risk_results['Late Delivery Probability'].apply(classify_risk)

risk_results['Risk Category'].value_counts()
```

```python
risk_results['Actual Late Delivery'] = y_test.values

risk_results.groupby('Risk Category')['Actual Late Delivery'].mean()
```

```python
import shap

explainer = shap.TreeExplainer(xgb_model)
shap_explanation = explainer(x_test)
```

```python
#global importance, but via SHAP
shap.summary_plot(shap_explanation.values, x_test)
```

```python
#Explaining High Risk Order
high_risk_orders = risk_results[risk_results['Risk Category'] == 'High Risk']
sample_idx = high_risk_orders.index[0]
pos = x_test.index.get_loc(sample_idx)

shap.plots.waterfall(shap_explanation[pos])
```

```python
#Plain Language Output
def explain_order(order_position, top_n=5):
    contributions = pd.DataFrame({
        'Feature': x_test.columns,
        'SHAP Value': shap_explanation.values[order_position]
    })
    contributions['Direction'] = np.where(contributions['SHAP Value'] > 0, 'increases risk', 'decreases risk')
    contributions = contributions.reindex(
        contributions['SHAP Value'].abs().sort_values(ascending=False).index
    )
    prob = best_proba[order_position]
    print(f'Predicted late-delivery probability: {prob:.1%}')
    print(f'\nTop {top_n} contributing factors:')
    for _, row in contributions.head(top_n).iterrows():
        print(f"  - {row['Feature']}: {row['Direction']} (impact: {row['SHAP Value']:.3f})")
    
explain_order(pos)
```

```python
#Key Risk Drivers
def get_top_drivers(shap_vals, feature_names, top_n=3):
    top_drivers = []
    for row in shap_vals:
        idx = np.argsort(np.abs(row))[::-1][:top_n]
        drivers = [feature_names[i] for i in idx]
        top_drivers.append(','.join(drivers))
    return top_drivers

risk_results['Key Risk Drivers'] = get_top_drivers(shap_explanation.values, x_test.columns, top_n=3)

risk_results.head()
```
